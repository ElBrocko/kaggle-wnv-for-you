```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import os
import numpy as np
import math
from sklearn import preprocessing
from sklearn.decomposition import PCA 

lean_week = pd.read_csv('train_weather_week_lean.csv')
```

```python
sns.set()
sns.set_context("paper")
plt.figure(figsize=(14, 14))
sns.heatmap(lean_week.corr(), cmap="RdBu")
```

```python
lean_week.dtypes
```

```python
#Were going to run the PCA on this df
lean_week_drop = lean_week.drop(['start date', 'Unnamed: 0', 'Mosquitos', 'WNV_Present'],
                                 axis=1)
#We're going to run a model on df with WNV_Present as the y
lean_week_clean = lean_week.drop(['start date', 'Unnamed: 0','Mosquitos'],
                                 axis=1)
```

```python
lean_week_clean_std =(lean_week_clean - lean_week_clean.mean())/lean_week_clean.std()

lean_week_drop_std = (lean_week_drop - lean_week_drop.mean())/lean_week_drop.std()
```

```python
lean_week_drop_std.head()
```

```python
lean_week_clean_std.head()
```

```python
lean_week_drop_std.plot(kind='line',legend=True, figsize=(12,10))
```

```python
X = np.cov(lean_week_drop_std)
sns.heatmap(np.cov(X), cmap="RdBu")
sns.set()
plt.figure(figsize=(20, 20))
```

```python
pca = PCA(n_components=2)
pca.fit(lean_week_drop_std.values)
```

```python
lean_week_drop_std.head()
```

```python
print pca.explained_variance_ratio_.sum()
```

```python
pca = PCA(n_components=5)
pca.fit(lean_week_drop_std)
```

```python
print pca.explained_variance_ratio_.sum()
```

Below we are doing the PCA with normalized dataframe

```python
lean_week_drop_norm = preprocessing.normalize(lean_week_drop, axis=0)
```

```python
pca = PCA(n_components=5)
pca.fit(lean_week_drop_norm)
```

```python
print pca.explained_variance_ratio_.sum()
```

Below we are using the min max scaler to standardize the dataframe 

```python
lean_week_clean_minmax = lean_week_clean.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
lean_week_drop_minmax = lean_week_drop.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
```

```python
lean_week_clean_minmax.head().describe()
```

```python
lean_week_drop_minmax.mean()
```

```python
lean_week_drop_cols = lean_week_drop.columns
```

```python
lean_week_drop_pca = PCA(n_components=5)
lean_week_drop_pca.fit(lean_week_drop_minmax)
```

```python
print lean_week_drop_pca.explained_variance_ratio_.sum()
```

```python
## Return the loadings for each of the components
lean_week_drop_components = lean_week_drop_pca.components_
```

```python
lean_week_drop_exp_var_eigenvals = lean_week_drop_pca.explained_variance_
lean_week_drop_exp_var_pct = lean_week_drop_pca.explained_variance_ratio_

print 'eigenvalues:', lean_week_drop_exp_var_eigenvals
print ''
print 'explained variance pct:', lean_week_drop_exp_var_pct
print ''
print lean_week_drop_exp_var_pct.sum()
```

```python
print lean_week_drop_cols
print '-------------------------------------'

for i, pc in enumerate(['PC1','PC2','PC3','PC4','PC5']):
    print pc, 'weighting vector:', lean_week_drop_components[i]
    print '-------------------------------------'
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline

pca = PCA()
lr = LinearRegression()

pca_pipe = make_pipeline(pca, lr)

pca_grid = {
    'pca__n_components':[1,2,3,4,5]
}

from sklearn.model_selection import GridSearchCV

# Gridsearch now takes the pipeline:
pca_gs = GridSearchCV(pca_pipe, pca_grid, cv=6)

# fit it on the data X, y as usual
pca_gs.fit(lean_week_drop_minmax, lean_week_clean_minmax['WNV_Present'])

print pca_gs.best_params_
print pca_gs.best_score_
```

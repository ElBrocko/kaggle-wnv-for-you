```python
# find tmax, tmin, tavg, dewpoint, heat, cool, preciptotal, stnpressure, resultdir, avgspeed from
# fifteen days prior

# Need to use try-except statements

weather_cols = ['Tmax','Tmin','Tavg','DewPoint',
                'Heat','Cool','PrecipTotal',
                'StnPressure','ResultDir','AvgSpeed']

#List of lists that will have all the 15 days prior weather data
new_cols = [weather_cols]
#iterate through every row
for i in df.index:
    new_row = [] #create a new, empty row
    #create a new datafram containing all rows that occurred 15 days before the ith row
    #and were at the same station
    fif_days_before = df[(df['Date'] == (df.loc[i].Date - pd.Timedelta('15 days'))) & 
                         (df['Station'] == df.loc[i].Station)]
    if len(fif_days_before) == 0:
        new_row = [None, None, None, None, None, None, None, None, None, None]
    else:
        for j in weather_cols:
            new_row.append(fif_days_before.iloc[0][j])
    new_cols.append(new_row)
```
